## Join nwt and cdc prior to interpolation of missing values

```python
nwt
```

```python
#Drop the domain columns in both tables, not needed
nwt = nwt.drop(columns=['domain'])
cdc = cdc.drop(columns=['domain'])
```

```python
#Group by siteid and date to get the mean for features that have multiple entries per date per site
nwt = nwt.groupby(['siteid', 'date'])[['nitrate_mean', 'spec_conductance', 'DO', 'sealevelDO_sat', 'pH', 
                               'chlorophyll', 'turbidity', 'fDOM', 'spectrum_count', 
                               'mean_temp']].mean()
#Group by siteid and date to get the mean for features that have multiple entries per date per site
cdc = cdc.groupby(['siteid', 'date'])[['CH4_conc', 'CO2_conc', 'N2O_conc', 'Microbialabundanceper_ml']].mean()
```

```python
nwt
```

```python
nwt = nwt.reset_index()
cdc = cdc.reset_index()
```

```python
cdc.info()
```

```python
#convert date dtype to datetime
nwt['date'] = pd.to_datetime(nwt['date'])
cdc['date'] = pd.to_datetime(cdc['date'])

#create a year_month column in both tables
nwt['year_month'] = nwt['date'].dt.to_period('M')
cdc['year_month'] = cdc['date'].dt.to_period('M')
```

```python
nwt.info()
```

```python
#merge nwt and cdc on siteid, date, and year_month
ft = nwt.merge(cdc, on =['siteid','date','year_month'], how='left')
ft
```

```python
import seaborn as sns
import matplotlib as plt
from ydata_profiling import ProfileReport

#generate profile report
ft_profile = ProfileReport(ft)
ft_profile.to_file("Full_table_preimpute..html")
```

```python
ft.info()
```

```python
#Drop spectrum count, feature is a measure of fDOM and chlorophyll measurements 
#Drop sealevelDO since it is a transformation of DO
ft = ft.drop(columns=['spectrum_count','sealevelDO_sat'])
```

```python
#pH is between 0 and 14, convert any nonsensical values to later be interpolated
pHcond = (ft['pH'] < 0) | (ft['pH'] > 14)
ft['pH'] = np.where(pHcond, np.NaN, ft['pH'])

#chlorophyll is measured as micrograms per L, can't be negative, coerce negative values to 0
chlcond = ft['chlorophyll'] < 0
ft['chlorophyll'] = np.where(chlcond, 0, ft['chlorophyll'])

#turbidity is between 0 and 100 in NTU(Nephelometric Turbidity Units), set negative values and values greater than 100 to NaN for interpolation
turbcond = (ft['turbidity'] < 0) | (ft['turbidity'] > 100)
ft['turbidity'] = np.where(turbcond, np.NaN, ft['turbidity'])

#Surface water mean temp is between 1.9 and 32 degrees Celsius, set outliers to NaN for interpolation
tempcond = (ft['mean_temp'] < 1.9) | (ft['mean_temp'] > 32)
ft['mean_temp'] = np.where(tempcond, np.NaN, ft['mean_temp'])

#Coerce negative values of DO to 0
DOcond = ft['DO'] < 0
ft['DO'] = np.where(DOcond, 0, ft['DO'])

#Nitrate mean concentration is 0 or greater micromol per liter, coerce negative values to 0
Nitcond = ft['nitrate_mean'] < 0
ft['nitrate_mean'] = np.where(Nitcond, 0, ft['nitrate_mean'])

#fluorescent dissolved organic matter is measured in quinine sulfate units which is between 0 and 300, coerce extremes to NaN for interpolation
fdomcond = (ft['fDOM'] < 0) | (ft['fDOM'] > 300)
ft['fDOM'] = np.where(fdomcond, np.NaN, ft['fDOM'])

#specific conductance is measured in microsiemens per cm and is greater than 0, coerce negatives to NaN for interpolation
spccond = ft['spec_conductance'] < 0
ft['spec_conductance'] = np.where(spccond, np.NaN, ft['spec_conductance'])

ft
```

```python
tmp = ft.groupby(['siteid', 'year_month'])['CH4_conc', 'CO2_conc', 'N2O_conc', 'Microbialabundanceper_ml'].apply(lambda x: x.mean())
ft = ft.merge(tmp, on=['siteid', 'year_month'], how='right')
ft
```

```python
#convert data to datetime datatype and rename CH4, CO2, N2O, and microbialabundance columns

ft = ft.drop(['CH4_conc_x','CO2_conc_x','N2O_conc_x', 'Microbialabundanceper_ml_x'], axis = 1)
ft = ft.rename(columns={'CH4_conc_y': 'CH4_conc', 'CO2_conc_y': 'CO2_conc', 'N2O_conc_y': 'N2O_conc', 'Microbialabundanceper_ml_y': 'Microbialabundanceper_ml'})
ft['date']= pd.to_datetime(ft['date'])
```

```python
#new columns to record rows with NaN 
ft['nitrate_mean_missing'] = ft['nitrate_mean'].isna()
ft['spec_conductance_missing'] = ft['spec_conductance'].isna()
ft['DO_missing'] = ft['DO'].isna()
ft['pH_missing'] = ft['pH'].isna()
ft['chlorophyll_missing'] = ft['chlorophyll'].isna()
ft['turbidity_missing'] = ft['turbidity'].isna()
ft['fDOM_missing'] = ft['fDOM'].isna()
ft['mean_temp_missing'] = ft['mean_temp'].isna()
ft['CH4_conc_missing'] = ft['CH4_conc'].isna()
ft['CO2_conc_missing'] = ft['CO2_conc'].isna()
ft['N2O_conc_missing'] = ft['N2O_conc'].isna()
ft['Microbialabundanceper_ml_missing'] = ft['Microbialabundanceper_ml'].isna()

#set date as index
ft.set_index('date', inplace = True)
ft
```

```python
#interpolate values for NaN using time method 
ft['nitrate_mean_missing'] = ft['nitrate_mean'].interpolate(method='time')
ft['spec_conductance_missing'] = ft['spec_conductance'].interpolate(method='time')
ft['DO_missing'] = ft['DO'].interpolate(method='time')
ft['pH_missing'] = ft['pH'].interpolate(method='time')
ft['chlorophyll_missing'] = ft['chlorophyll'].interpolate(method='time')
ft['turbidity_missing'] = ft['turbidity'].interpolate(method='time')
ft['fDOM_missing'] = ft['fDOM'].interpolate(method='time')
ft['mean_temp_missing'] = ft['mean_temp'].interpolate(method='time')
ft['CH4_conc_missing'] = ft['CH4_conc'].interpolate(method='time')
ft['CO2_conc_missing'] = ft['CO2_conc'].interpolate(method='time')
ft['N2O_conc_missing'] = ft['N2O_conc'].interpolate(method='time')
ft['Microbialabundanceper_ml_missing'] = ft['Microbialabundanceper_ml'].interpolate(method='time')
ft
```

```python
#assign values to new dataframe while renaming columns
new_df= ft[['siteid','nitrate_mean_missing', 'spec_conductance_missing', 'DO_missing', 
                          'pH_missing', 'chlorophyll_missing', 'turbidity_missing', 'fDOM_missing', 
                          'mean_temp_missing', 'CH4_conc_missing', 'CO2_conc_missing', 'N2O_conc_missing', 
                           'Microbialabundanceper_ml_missing']].rename(columns={'nitrate_mean_missing':'nitrate_mean', 
                          'spec_conductance_missing':'spec_conductance', 'DO_missing':'DO', 
                          'pH_missing': 'pH', 'chlorophyll_missing':'chlorophyll', 'turbidity_missing':'turbidity', 'fDOM_missing':'fDOM', 
                          'mean_temp_missing':'mean_temp', 'CH4_conc_missing':'CH4_conc', 
                          'CO2_conc_missing':'CO2_conc','N2O_conc_missing':'N2O_conc', 'Microbialabundanceper_ml_missing':'Microbialabundanceper_ml' } 
                        )
new_df
```

## Create a log transformation of columns that show a log normal distribution for log-level and log-log ML models
